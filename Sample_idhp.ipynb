```python
import os
os.environ["TOKENIZERS_PARALLELISM"] = "false"
from transformers import AutoTokenizer, AutoModelForCausalLM
#Chang sample model here
model_name = "Qwen/Qwen3-0.6B"

sample_tokenizer = AutoTokenizer.from_pretrained(model_name)
sample_tokenizer.pad_token = sample_tokenizer.eos_token

sample_model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",  
    torch_dtype="auto"   
)
```

```python
import json
import numpy as np
import torch
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from Automation.new_sdscm_2 import format_sequences_as_dataframe, sample_counterfactual_sequences
from Automation.new_sdscm import sample_sequences
from Automation.estimation import run_estimation_methods_by_varname
import os

output_dir = "./new_data"
os.makedirs(output_dir, exist_ok=True)
os.makedirs(f"{output_dir}/plots", exist_ok=True)

np.random.seed(1)
torch.manual_seed(1)

# -----------------------
# Load candidate set JSON file
# -----------------------
def load_candidates_file(path):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
    return data

# -----------------------
# Evaluation Functions
# -----------------------
def calculate_pehe(estimated_cate, true_cate):
    """Precision in Estimation of Heterogeneous Effects"""
    return np.sqrt(np.mean((estimated_cate - true_cate) ** 2))

def check_positivity_violations(propensity_scores, threshold_low=0.1, threshold_high=0.9):
    """Check for positivity/overlap violations"""
    violations_low = np.sum(propensity_scores < threshold_low)
    violations_high = np.sum(propensity_scores > threshold_high)
    total_violations = violations_low + violations_high
    violation_rate = total_violations / len(propensity_scores)
    
    return {
        'violations_low': violations_low,
        'violations_high': violations_high,
        'total_violations': total_violations,
        'violation_rate': violation_rate,
        'min_propensity': np.min(propensity_scores),
        'max_propensity': np.max(propensity_scores)
    }

def calculate_propensity_scores(z_data):
    """Calculate empirical propensity scores"""
    propensity = np.mean(z_data)
    propensity_scores = np.full(len(z_data), propensity)
    return propensity_scores

def plot_propensity_distribution(propensity_scores, treatment, output_path):
    """Plot propensity score distribution by treatment group"""
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    
    # Histogram
    if len(propensity_scores[treatment == 0]) > 0:
        axes[0].hist(propensity_scores[treatment == 0], bins=30, alpha=0.5, label='Control (T=0)', density=True)
    if len(propensity_scores[treatment == 1]) > 0:
        axes[0].hist(propensity_scores[treatment == 1], bins=30, alpha=0.5, label='Treated (T=1)', density=True)
    axes[0].axvline(0.1, color='r', linestyle='--', label='Positivity threshold')
    axes[0].axvline(0.9, color='r', linestyle='--')
    axes[0].set_xlabel('Propensity Score')
    axes[0].set_ylabel('Density')
    axes[0].set_title('Propensity Score Distribution')
    axes[0].legend()
    
    # Box plot
    df_plot = pd.DataFrame({
        'Propensity': propensity_scores,
        'Treatment': ['Treated' if t == 1 else 'Control' for t in treatment]
    })
    sns.boxplot(data=df_plot, x='Treatment', y='Propensity', ax=axes[1])
    axes[1].axhline(0.1, color='r', linestyle='--')
    axes[1].axhline(0.9, color='r', linestyle='--')
    axes[1].set_title('Propensity Score by Treatment Group')
    
    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()

def plot_cate_distribution(estimated_cate, true_cate, method_name, output_path):
    """Plot CATE distribution and comparison with true CATE (robust version)"""
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    # ---- dynamic bins ----
    def choose_bins(arr):
        uniq = len(np.unique(arr))
        if uniq <= 1:
            return 1
        return min(uniq, 30)

    bins_est = choose_bins(estimated_cate)
    bins_true = choose_bins(true_cate)
    bins_err = choose_bins(estimated_cate - true_cate)

    # 1. histogram
    axes[0, 0].hist(estimated_cate, bins=bins_est, alpha=0.5,
                    label='Estimated CATE', density=True)
    axes[0, 0].hist(true_cate, bins=bins_true, alpha=0.5,
                    label='True CATE', density=True)

    axes[0, 0].axvline(np.mean(estimated_cate), color='blue', linestyle='--',
                       label=f'Est. ATE={np.mean(estimated_cate):.2f}')
    axes[0, 0].axvline(np.mean(true_cate), color='orange', linestyle='--',
                       label=f'True ATE={np.mean(true_cate):.2f}')
    axes[0, 0].set_title(f'{method_name}: CATE Distribution')
    axes[0, 0].legend()

    # 2. scatter
    axes[0, 1].scatter(true_cate, estimated_cate, alpha=0.5)
    min_val = min(true_cate.min(), estimated_cate.min())
    max_val = max(true_cate.max(), estimated_cate.max())
    axes[0, 1].plot([min_val, max_val], [min_val, max_val], 'r--')
    axes[0, 1].set_title(f'{method_name}: Estimated vs True CATE')

    # 3. error histogram
    errors = estimated_cate - true_cate
    axes[1, 0].hist(errors, bins=bins_err, edgecolor='black')
    axes[1, 0].axvline(0, color='red', linestyle='--')
    axes[1, 0].set_title(f'{method_name}: CATE Estimation Error')

    # 4. QQ plot
    from scipy import stats
    stats.probplot(errors, dist="norm", plot=axes[1, 1])
    axes[1, 1].set_title(f'{method_name}: Error Q-Q Plot')

    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    plt.close()

def calculate_ate_coverage(ate_lower, ate_upper, true_ate):
    """Check if confidence interval covers true ATE"""
    if ate_lower is None or ate_upper is None:
        return None
    return (ate_lower <= true_ate <= ate_upper)

# -----------------------
# Settings
# -----------------------
nums = [1,2,9]
types = ["long", "med", "short"]
num_samples = 672

METHOD_NAMES = [
    'CausalForest',
    'NaiveLinReg',
    'NaiveRF',
    'LinReg',
    'RF',
    'CausalForestDML',
    'LinearDML',
    'ForestDR',
    'LinearDR',
    'ForestTLearner',
    'ForestSLearner',
    'LinearTLearner',
    'LinearSLearner'
]

all_dataset_paths = []
dataset_metadata = {}

# =============================================================
# Step 1: Generate data
# =============================================================
print("="*60)
print("Step 1: Generating datasets with counterfactual data")
print("="*60)

for x in nums:
    for t in types:
        json_path = f"./DAGs_fixed/ihdp_complete_{x}_{t}_FIXED.json"
        ihdp_space = load_candidates_file(json_path)
        ihdp_samples_space = ihdp_space
        name_of_intervention = "t"
        name_of_outcome = "yf"

        all_covariates = [
            setup_dict['variable_name']
            for setup_dict in ihdp_space["setup_sequence_sample_space"]
        ]
        
        ihdp_samples = sample_sequences(
            model=sample_model,
            tokenizer=sample_tokenizer,
            sequence_sample_space=ihdp_samples_space,
            num_samples=num_samples,
        )

        df_samples_ihdp = format_sequences_as_dataframe(
            sequences=ihdp_samples,
            name_of_intervention=name_of_intervention,
            name_of_outcome=name_of_outcome,
            save_logprobs=True
        )
        
        print(f"\nDataset: {x}_{t}")
        print(f"  Total samples: {len(df_samples_ihdp)}")
        print(f"  Treatment distribution: {df_samples_ihdp[name_of_intervention].value_counts().to_dict()}")
        
        treatment_classes = df_samples_ihdp[name_of_intervention].nunique()
        if treatment_classes < 2:
            print(f"  WARNING: Only {treatment_classes} treatment class. Skipping.")
            continue

        np.random.seed(3)
        torch.manual_seed(3)

        index_of_intervention = np.where(
            [setup_dict['variable_name'] == name_of_intervention 
             for setup_dict in ihdp_space["setup_sequence_sample_space"]]
        )[0][0]
        
        evidence_indices = list(range(0, 15)) + [17]

        idhp_cf_samples_do_0 = sample_counterfactual_sequences(
            model=sample_model,
            tokenizer=sample_tokenizer,
            sequence_sample_space=ihdp_samples_space,
            samples=ihdp_samples,
            evidence_indices=evidence_indices,
            index_of_intervention=index_of_intervention,
            intervention_choice=0
        )
        
        idhp_cf_samples_do_1 = sample_counterfactual_sequences(
            model=sample_model,
            tokenizer=sample_tokenizer,
            sequence_sample_space=ihdp_samples_space,
            samples=ihdp_samples,
            evidence_indices=evidence_indices,
            index_of_intervention=index_of_intervention,
            intervention_choice=1
        )

        df_cf_do_0 = format_sequences_as_dataframe(
            sequences=idhp_cf_samples_do_0,
            name_of_intervention=name_of_intervention,
            name_of_outcome=name_of_outcome
        )

        df_cf_do_1 = format_sequences_as_dataframe(
            sequences=idhp_cf_samples_do_1,
            name_of_intervention=name_of_intervention,
            name_of_outcome=name_of_outcome
        )
        
        # Add counterfactual columns
        df_samples_ihdp[f'{name_of_outcome}|do({name_of_intervention}=0)'] = df_cf_do_0[name_of_outcome].values
        df_samples_ihdp[f'{name_of_outcome}|do({name_of_intervention}=1)'] = df_cf_do_1[name_of_outcome].values
        
        full_csv_name = f"{output_dir}/samples_{x}_{t}_full.csv"
        df_samples_ihdp.to_csv(full_csv_name, index=False)
        print(f"  Saved: {full_csv_name}")
        
        all_dataset_paths.append(full_csv_name)
        
        covariate_names = [col for col in all_covariates 
                          if col not in [name_of_intervention, name_of_outcome]]
        
        dataset_key = f"{x}_{t}"
        dataset_metadata[dataset_key] = {
            'space_num': x,
            'space_type': t,
            'csv_path': full_csv_name,
            'covariate_names': covariate_names,
            'treatment_name': name_of_intervention,
            'outcome_name': name_of_outcome,
            'treatment_index': index_of_intervention,
            'df': df_samples_ihdp
        }

# =============================================================
# Step 2: Run estimation methods
# =============================================================
if not all_dataset_paths:
    print("No valid datasets generated. Exiting.")
    exit()

print(f"\n{'='*60}")
print(f"Step 2: Running estimation methods")
print(f"{'='*60}\n")

first_key = list(dataset_metadata.keys())[0]
first_meta = dataset_metadata[first_key]
covariate_names = first_meta['covariate_names']
treatment_name = first_meta['treatment_name']
y_name = first_meta['outcome_name']


first_df = pd.read_csv(first_meta['csv_path'])

treatment_index = first_meta['treatment_index']
print(f"Treatment index (from metadata): {treatment_index}")


all_columns = list(first_df.columns)
outcome_index = all_columns.index(y_name)
print(f"Outcome index: {outcome_index} (column: {y_name})")

possible_treatment_values = sorted(first_df[treatment_name].unique().tolist())
print(f"Possible treatment values: {possible_treatment_values}")


possible_outcome_values = sorted(first_df[y_name].unique().tolist())
print(f"Possible outcome values: {possible_outcome_values[:10]}..." if len(possible_outcome_values) > 10 else f"Possible outcome values: {possible_outcome_values}")

has_propensity = any('logP' in col for col in first_df.columns)
print(f"Has propensity scores: {has_propensity}")

if not has_propensity:
    print("\nNote: No propensity scores found in data. The add_normalized_colums function will skip propensity normalization.")

print(f"\n{'='*60}")
print("Running estimation methods with parameters:")
print(f"  treatment_index: {treatment_index}")
print(f"  outcome_index: {outcome_index}")
print(f"  possible_treatment_values: {possible_treatment_values}")
print(f"  # of possible_outcome_values: {len(possible_outcome_values)}")
print(f"{'='*60}\n")

results_df = run_estimation_methods_by_varname(
    dataset_paths=all_dataset_paths,
    method_names=METHOD_NAMES,
    covariate_names=covariate_names,
    treatment_name=treatment_name,
    y_name=y_name,
    use_normalization=False,
    num_processes=4
)

raw_results_path = f"{output_dir}/all_estimation_results_raw.pkl"
pd.to_pickle(results_df, raw_results_path)
print(f"\nSaved raw results: {raw_results_path}")

# =============================================================
# Step 3: Evaluation
# =============================================================
print(f"\n{'='*60}")
print("Step 3: Calculating evaluation metrics")
print(f"{'='*60}\n")

evaluation_results = []
positivity_results = []

for idx, row in results_df.iterrows():
    df_path = row['df_path']
    method_name = row['method_name']
    
    filename = os.path.basename(df_path)
    parts = filename.replace('samples_', '').replace('_full.csv', '').split('_')
    dataset_key = f"{parts[0]}_{parts[1]}"
    
    if dataset_key not in dataset_metadata:
        continue
    
    meta = dataset_metadata[dataset_key]
    
    print(f"Processing: {dataset_key} - {method_name}")
    
    y0_true = row['y0']
    y1_true = row['y1']
    true_cate = y1_true - y0_true
    true_ate = np.mean(true_cate)
    
    z_data = row['treatment']
    y_data = row['y']
    
    estimated_ate = row['ate']
    estimated_cate = row['cate']
    ate_lower = row['ate_lower_bound']
    ate_upper = row['ate_upper_bound']
    
    ate_bias = estimated_ate - true_ate
    ate_abs_bias = abs(ate_bias)
    ate_coverage = calculate_ate_coverage(ate_lower, ate_upper, true_ate)
    
    estimated_cate = np.asarray(estimated_cate).reshape(-1)
    true_cate = np.asarray(true_cate).reshape(-1)

    if estimated_cate is not None and len(estimated_cate) == len(true_cate):
        pehe = calculate_pehe(estimated_cate, true_cate)
        cate_rmse = pehe
        cate_correlation = np.corrcoef(estimated_cate, true_cate)[0, 1]
        true_cate_std = np.std(true_cate)
        estimated_cate_std = np.std(estimated_cate)
        cate_std_ratio = estimated_cate_std / true_cate_std if true_cate_std > 0 else None
        cate_mae = np.mean(np.abs(estimated_cate - true_cate))
        cate_variance = np.var(true_cate)
        ate_captures_heterogeneity = (cate_variance > 0.01)
        
        plot_path = f"{output_dir}/plots/cate_{dataset_key}_{method_name}.png"
        try:
            plot_cate_distribution(estimated_cate, true_cate, method_name, plot_path)
        except Exception as e:
            print(f"  Warning: Could not generate CATE plot: {e}")
    else:
        pehe = None
        cate_rmse = None
        cate_correlation = None
        cate_std_ratio = None
        cate_mae = None
        cate_variance = np.var(true_cate)
        ate_captures_heterogeneity = (cate_variance > 0.01)
        estimated_cate_std = None
    
    propensity_scores = calculate_propensity_scores(z_data)
    positivity_violations = check_positivity_violations(propensity_scores)
    
    propensity_plot_path = f"{output_dir}/plots/propensity_{dataset_key}.png"
    if not os.path.exists(propensity_plot_path):
        try:
            plot_propensity_distribution(propensity_scores, z_data, propensity_plot_path)
        except Exception as e:
            print(f"  Warning: Could not generate propensity plot: {e}")
    
    positivity_results.append({
        'space_num': meta['space_num'],
        'space_type': meta['space_type'],
        'dataset_key': dataset_key,
        **positivity_violations
    })
    
    eval_row = {
        'space_num': meta['space_num'],
        'space_type': meta['space_type'],
        'dataset_key': dataset_key,
        'method_name': method_name,
        'dataset_size': row['dataset_size'],
        'estimated_ate': estimated_ate,
        'true_ate': true_ate,
        'ate_bias': ate_bias,
        'ate_abs_bias': ate_abs_bias,
        'ate_lower_bound': ate_lower,
        'ate_upper_bound': ate_upper,
        'ate_coverage': ate_coverage,
        'pehe': pehe,
        'cate_rmse': cate_rmse,
        'cate_mae': cate_mae,
        'cate_correlation': cate_correlation,
        'true_cate_std': np.std(true_cate),
        'estimated_cate_std': estimated_cate_std,
        'cate_std_ratio': cate_std_ratio,
        'true_cate_variance': cate_variance,
        'ate_captures_heterogeneity': ate_captures_heterogeneity,
        'positivity_violation_rate': positivity_violations['violation_rate'],
        'positivity_violations_total': positivity_violations['total_violations'],
    }
    
    evaluation_results.append(eval_row)
    print(f"  ATE Bias: {ate_bias:.4f}, PEHE: {f'{pehe:.4f}' if pehe is not None else 'N/A'}")


# =============================================================
# Step 4: Save results
# =============================================================
print(f"\n{'='*60}")
print("Step 4: Saving results")
print(f"{'='*60}\n")

eval_df = pd.DataFrame(evaluation_results)
eval_results_path = f"{output_dir}/comprehensive_evaluation_results.csv"
eval_df.to_csv(eval_results_path, index=False)
print(f"Saved: {eval_results_path}")

positivity_df = pd.DataFrame(positivity_results).drop_duplicates(subset=['dataset_key'])
positivity_path = f"{output_dir}/positivity_analysis.csv"
positivity_df.to_csv(positivity_path, index=False)
print(f"Saved: {positivity_path}")

print("\n--- Performance by Method ---")
method_summary = eval_df.groupby('method_name').agg({
    'ate_abs_bias': ['mean', 'std', 'min', 'max'],
    'pehe': ['mean', 'std', 'min', 'max'],
    'cate_correlation': ['mean', 'std'],
    'ate_coverage': 'mean'
}).round(4)
print(method_summary)

method_summary_path = f"{output_dir}/method_summary_statistics.csv"
method_summary.to_csv(method_summary_path)
print(f"Saved: {method_summary_path}")

print("\n--- Performance by Dataset ---")
dataset_summary = eval_df.groupby('dataset_key').agg({
    'ate_abs_bias': ['mean', 'std'],
    'pehe': ['mean', 'std'],
    'true_cate_variance': 'first',
    'positivity_violation_rate': 'first'
}).round(4)
print(dataset_summary)

dataset_summary_path = f"{output_dir}/dataset_summary_statistics.csv"
dataset_summary.to_csv(dataset_summary_path)
print(f"Saved: {dataset_summary_path}")

# Best methods
print("\n--- Top 5 Methods by Metric ---")
metrics_of_interest = ['ate_abs_bias', 'pehe', 'cate_correlation']
for metric in metrics_of_interest:
    if metric in eval_df.columns and eval_df[metric].notna().any():
        if metric == 'cate_correlation':
            best = eval_df.nlargest(5, metric)[['method_name', 'dataset_key', metric]]
        else:
            best = eval_df.nsmallest(5, metric)[['method_name', 'dataset_key', metric]]
        print(f"\nBest by {metric}:")
        print(best.to_string(index=False))

print("\n" + "="*60)
print("All processing complete!")
print(f"Results in: {output_dir}")
print(f"Plots in: {output_dir}/plots")
print("="*60)
```
